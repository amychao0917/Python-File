## Google 翻譯
可以用一個套件，叫做
googletrans

```python
!pip install googletrans 
```

```python
from googletrans import Translator
```

```python
pipi = Translator()
```

```python
result = pipi.translate("This is a book.", dest="zh-tw")
```

```python
print(result.text)
```

### 基本指令

* 'src':原文
* 'dest':目標語言

常用語言:
*中文: zh-tw
*日文:ja
*英文:en
*韓文:ko

## 'request'+'BeautifulSoup'

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.marieclaire.com.tw/love-sex/astrology"
```

```python
resp = requests.get(url)
```

```python
print(resp.text)
```

```python
soup = BeautifulSoup(resp.text, 'lxml')
```

```python
hot = soup.find_all('div', class_="hot")
```

```python
articles = hot[0].find_all('a')
```

```python
len(articles)
```

```python
item = articles[2]
```

```python
item.attrs
```

```python
item['href']
```

```python
title = item.find('h3')
```

```python
title.contents[0]
```
