## 麻州 8 年級學生數學程度

```python
df1 = pd.read_csv("states.csv")

df1_math1 = df1[["STATE","YEAR","AVG_MATH_8_SCORE"]]
df1_math1.sort_values(by = ["AVG_MATH_8_SCORE"], ascending= False, inplace= True)
df1_math1.index= range(1,len(df1_math1)+1)
```

```python
df1_MAmath = df1_math1.drop(df1_math1[df1_math1.STATE!= "MASSACHUSETTS"].index)
```

```python
df1_MAmath
```
