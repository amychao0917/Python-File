```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
```

```python
ax = Axes3D(fig) #生成3D物件
X = np.arange(-4,4,0.25)
Y = np.arange(-4,4,0.25)
X,Y = np.meshgrid(X,Y)

R = np.sqrt(X**2+Y**2)
Z = np.sin(R)
```

```python
ax.set_xlabel('X label',color='g')
ax.set_ylabel('Y label',color='g')
ax.set_zlabel('Z label')
ax.plot_surface(X,Y,Z,rstride = 1,cstride = 1, cmap='twilight') #生成曲面
ax.contour(X,Y,Z, offset = -2, colors='black') #生成等高線，offset引數是等高線線所處位置
```

```python
fig = plt.figure()

bx = fig.add_subplot(111)

C= bx.contour(X,Y,Z)
bx.clabel(C,inline=True,fontsize=10)
ax.set_zlim(-2,2)
plt.show()
```
