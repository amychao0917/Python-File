```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/chipotle.tsv",delimiter="\t")
```

```python
df.head(10)
```

```python
df_item = df.groupby("item_name").count()
```

```python
df_item.sort_values(by="quantity",ascending = False)
```

```python
df_item = df_item[df_item.quantity > 100]
df_item.quantity.plot(kind = "bar")
```
