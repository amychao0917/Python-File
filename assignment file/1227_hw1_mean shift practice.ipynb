```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Mean Shift

```python
x = np.random.randn(100,2) 
```

```python
from sklearn.cluster import MeanShift
```

```python
clf = MeanShift(bandwidth=0.5)
```

```python
clf.fit(x)
```

```python
clf.labels_
```

```python
plt.scatter(x[:,0],x[:,1], c=clf.labels_, cmap='tab20')
```

```python
x0 = np.arange(-3,4,0.02)
y0 = np.arange(-1,3,0.02)

xm, ym = np.meshgrid(x0,y0)

P = np.c_[xm.ravel(),ym.ravel()]
z = clf.predict(P)
Z = z.reshape(xm.shape)

plt.contourf(xm, ym, Z, alpha=0.3, cmap='tab20')
plt.scatter(x[:,0],x[:,1],c=clf.labels_, cmap='tab20')
```
