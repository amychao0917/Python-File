```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def height(x,y):
    return(1-x/2+x**5+y**3)*np.exp(-x**2-y**2) #exp指數函數
```

```python
n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
```

```python
X,Y = np.meshgrid(x,y)
```

```python
plt.contourf(X, Y, height(X,Y),8,alpha = 0.8, cmap='afmhot') #8代表等高線圖分10份
contour_lines = plt.contour(X,Y,height(X,Y), 8, color='black',lw = 0.5)
plt.clabel(contour_lines, inline=True, fontsize=10)
```
